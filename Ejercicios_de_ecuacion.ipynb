```python
#Ejercicio 1'
import numpy as np
import matplotlib.pyplot as plt

# problema a
a1 = np.array([[2,1],[-3,3]])
b1 = np.array([1,5])
x1 = np.linalg.solve(a1,b1)
print(f"Respuesta: {x1}")

# interseccion de dos lineas
x_vals = np.linspace(-5, 5, 100)
y1_eq1 = (1 - 2*x_vals) / 1
y1_eq2 = (5 + 3*x_vals) / 3

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y1_eq1, label='2x + y = 1')
plt.plot(x_vals, y1_eq2, label='-3x + 3y = 5')
plt.plot(x1[0], x1[1], 'ro', label=f'Solucion: ({x1[0]:.2f}, {x1[1]:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico del problema a')
plt.grid(True)
plt.legend()
plt.show()


# problema b
a2 = np.array([[5,2],[4,3]])
b2 = np.array([6,14])
x2 = np.linalg.solve(a2,b2)
print(f"Respuesta: {x2}")

# Interseccion de dos lineas
y2_eq1 = (6 - 5*x_vals) / 2
y2_eq2 = (14 - 4*x_vals) / 3

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y2_eq1, label='5x + 2y = 6')
plt.plot(x_vals, y2_eq2, label='4x + 3y = 14')
plt.plot(x2[0], x2[1], 'ro', label=f'Solucion: ({x2[0]:.2f}, {x2[1]:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafico del problema b')
plt.grid(True)
plt.legend()
plt.show()
```

 Sistema de ecuaciones lineales
 x + 2y = 1
 3x + 5y = 2

```python
import numpy as np
a = np.array([[1,2],[3,5]])
b = np.array([1,2])

#np.linag.solve(m1, m2)

x = np.linalg.solve(a,b)

print(x)
```

```python
import numpy as np
a = np.array([[1,6],[7,-3]])
b = np.array([27,9])

#np.linag.solve(m1, m2)

x = np.linalg.solve(a,b)

print(x)
```

```python
import numpy as np
a = np.array([[1,1,2],[3,1,-2],[5,-2,5]])
b = np.array([8,10,12])

#np.linag.solve(m1, m2)

x = np.linalg.solve(a,b)
print(np.round(x,2))
```

Ecuaciones lineales con artificios


```python
import numpy as np
#registrar variables
A=np.matrix([[1,2,1],[2,2,1],[3,3,5]])
B=np.matrix([[8],[9],[24]])

#primer metodo
print('Primera solucion')
x=(A**-1) @ B
print(x)

#Segundo metodo
print('Segunda solucion')
y=np.linalg.solve(A,B)
print(y)
```

Ecuaciones cuadraticas con estadisticos

```python
import matplotlib.pyplot as plt
import numpy as np

#1. definir los datos
x = np.linspace(-5,5,100)#valores de x
y = x**2 #valores de y para y = x²

#2. crear el grafico
plt.figure(figsize=(10,5))#tamaño del grafico
plt.plot(x,y,color='purple', label='y = x²')#trasar linea del grafico

#3. Añadir elementos al grafico
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Visualisacion de la ecuacion y=x²')
plt.legend()
plt.grid(True)

#4. Mostrar el grafico
plt.show()
```
